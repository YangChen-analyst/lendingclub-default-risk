```python
import pandas as pd
import numpy as np
```

```python
import os
print(os.getcwd())
df = pd.read_csv('../data/lending_club_loan_two.csv')
```

```python
print('Shape: ', df.shape)
```

```python
print('Info: ', df.info())
```

```python
df.head()
```

```python
# Corrected way to remove duplicates
df.drop_duplicates(inplace=True)

# Now drop rows with missing values exceeding 30%
df.dropna(thresh=df.shape[1]*0.7, axis=0, inplace=True)

# Forward-fill remaining missing values
df.fillna(method='ffill', inplace=True)
```

```python
# Create Default flag
df['default_flag'] = df['loan_status'].apply(lambda x : 1 if x == 'Charged Off' else 0)
df['default_flag'].value_counts(normalize=True)
# Extract State
df['state'] = df['address'].str.extract(r',\s*([A-Z]{2})\s*\d{5}$')
# Drop address
df.drop('address', axis=1, inplace=True)
```

```python
# Export the DF to CSV
output_file_path = '../data/cleaned_data.csv'
df.to_csv(output_file_path, index=False)
```
